```python
def greet():
    print("Привет, Хабр!")

# Присваиваем функцию переменной. 
# ВАЖНО: мы пишем имя без скобок! Мы не вызываем функцию, мы ее передаем.
say_hello = greet 

# Теперь say_hello делает то же самое, что и greet
say_hello() 
# Вывод: Привет, Хабр!
```

```python
def shout():
    print("УРААА!")

def whisper():
    print("тссс...")

# Эта функция принимает ДРУГУЮ функцию в качестве аргумента
def execute_twice(func):
    func() # Вызываем переданную функцию первый раз
    func() # И второй раз

execute_twice(shout)
# Вывод:
# УРААА!
# УРААА!

execute_twice(whisper)
# Вывод:
# тссс...
# тссс...
```

## концепция замыкания (closure)

Внутренняя функция multiplier запомнила значение переменной n из внешней функции make_multiplier, даже когда внешняя функция уже завершила свою работу!

```python
def make_multiplier(n):
    # Создаем внутреннюю функцию
    def multiplier(x):
        return x * n
    
    # Возвращаем саму функцию, а не результат ее работы! (без скобок)
    return multiplier 

# Создаем функцию, которая всегда умножает на 3
times_three = make_multiplier(3)

print(times_three(10)) # Выведет 30
print(times_three(5))  # Выведет 15
```

## первый декоратор

```python
def polite_decorator(func):
    # Создаем функцию-обертку
    def wrapper():
        print("Здравствуйте! Начинаем работу.")
        
        func() # Вызываем оригинальную функцию
        
        print("Работа завершена. До свидания!")
    
    # Возвращаем обертку (без скобок!)
    return wrapper

def do_work():
    print("Пилю фичи, фиксим баги...")

# Передаем оригинальную функцию в декоратор и перезаписываем её же имя!
do_work = polite_decorator(do_work)

# Вызываем функцию
do_work()
```

```python
# синтаксический сахар @

def polite_decorator(func):
    def wrapper():
        print("Здравствуйте! Начинаем работу.")
        func()
        print("Работа завершена. До свидания!")
    return wrapper

# Используем синтаксический сахар
@polite_decorator
def do_work():
    print("Пилю фичи, фиксим баги...")

do_work()
```

# Аргументы и возврат значений (Как не сломать оригинальную функцию)

<code>
def wrapper(*args, **kwargs):
    # ... код до ...
    result = func(*args, **kwargs)
    # ... код после ...
    return result
</code>

```python
def polite_decorator(func):
    def wrapper(*args, **kwargs):
        print("Здравствуйте! Начинаем работу.")
        
        # Ловим результат работы оригинальной функции
        result = func(*args, **kwargs) 
        
        print("Работа завершена. До свидания!")
        
        # Честно отдаем его наружу
        return result 
        
    return wrapper

@polite_decorator
def add_numbers(a, b):
    return a + b

result = add_numbers(10, 20)
print(f"Результат сложения: {result}")
```

```python
# Сохранение метаданных функции с помощью functools.wraps
from functools import wraps

def polite_decorator(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        print("Здравствуйте! Начинаем работу.")
        
        # Ловим результат работы оригинальной функции
        result = func(*args, **kwargs) 
        
        print("Работа завершена. До свидания!")
        
        # Честно отдаем его наружу
        return result 
        
    return wrapper

@polite_decorator
def add_numbers(a, b):
    """Складывает два числа и возвращает результат."""
    return a + b

result = add_numbers(10, 20)
print(f"Результат сложения: {result}")


print(add_numbers.__name__)  # Вывод: add_numbers
print(add_numbers.__doc__)   # Вывод: None (или строка документации, если она была)
```
